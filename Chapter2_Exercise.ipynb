```python
import pandas as pd
pd.options.display.max_rows = 1000
import numpy as np
from pandas.tools.plotting import scatter_matrix
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline 
```

```python
# Exercise 8
# 8 (a) and 8 (b) reading data
df_college = pd.read_csv('Data/College.csv',index_col =0)
df_college.head()
```

```python
# 8 (c) i
df_college.describe(include='all')
```

```python
# 8 (c) ii
#df_college.iloc[:,0:10].head()
#sns.pairplot(df_college.iloc[:,0:5])
scatter_matrix(df_college.iloc[:,0:10],figsize=(10, 10))
# note the categorical variable gets lost as compared to R 
```

```python
# 8 c (iii)
# creating box plots
df_college.boxplot(column= ['Outstate'], by = ['Private'],figsize = (10,5))
```

```python
# 8(c) (iv)
df_college['Elite'] = ['Yes' if x > 50 else 'No' for x in df_college['Top10perc']]
# using list comprehension to get the best perfomance
print(df_college['Elite'].describe())
print(df_college['Elite'].value_counts())
df_college.boxplot(column= ['Outstate'], by = ['Elite'], figsize= (10,5))
```

```python
# 8 (c) (v) drawing multiplt histograms
fig, axes = plt.subplots(nrows=2, ncols=2)
ax0, ax1, ax2, ax3 = axes.flat

ax0.hist(df_college['Apps'], color= 'tan')
ax0.set_xlabel('Apps')
ax0.set_ylabel('Frequency')
ax0.set_title('Histogram of Apps')

ax1.hist(df_college['perc.alumni'], color= 'red')
ax1.set_xlabel('% of Alumni')
ax1.set_ylabel('Frequency')
ax1.set_title('Histogram of % of Alumni')



ax2.hist(df_college['S.F.Ratio'], color= 'blue', bins= 10)
ax2.set_xlabel('S.F ratio')
ax2.set_ylabel('Frequency')
ax2.set_title('Histogram of % of S.F ratio')


ax3.hist(df_college['Expend'], color= 'green', bins= 100)
ax3.set_xlabel('Expend')
ax3.set_ylabel('Frequency')
ax3.set_title('Histogram of % of Expend')

fig.set_figwidth(10)
fig.set_figheight(8)
plt.tight_layout()
plt.show()
```

```python
# 8 (vi)
fig, axes = plt.subplots(nrows=1, ncols=3)
ax0, ax1, ax2  = axes.flat

ax0.scatter(df_college['Outstate'], df_college['Grad.Rate'], color = 'tan')
ax0.set_xlabel('Outstate')
ax0.set_ylabel('Grad.Rate')
# High out of state tution --> high grad rate

ax1.scatter(df_college['Accept']/df_college['Apps'], df_college['S.F.Ratio'], color = 'blue')
ax1.set_xlabel('Acceptance Ratio')
ax1.set_ylabel('S.F Ratio')
# Low acceptance ration --> low S/F ratio

ax2.scatter(df_college['Top10perc'], df_college['Grad.Rate'], color = 'red')
ax2.set_xlabel('Top 10 perc')
ax2.set_ylabel('Grad.Rate')
# Colleges with the most students from top 10% perc don't necessarily have
# the highest graduation rate. Also, rate > 100 is erroneous!

fig.set_figwidth(12)
fig.set_figheight(5)
plt.tight_layout()
plt.show()
```
